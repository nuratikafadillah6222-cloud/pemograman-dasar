```python
# Soal 1

# Meminta input suhu dari pengguna
suhu = float(input("Suhu air pendingin dalam (°C): "))

# Mengecek kondisi suhu dengan percabangan
if suhu <= 90:
    print("STATUS: Normal")
elif suhu > 90 and suhu < 100:
    print("STATUS: Warning")
else:   # suhu >= 100
    print("ALERT: SHUTDOWN")
```

**2.   percabangan if, elif, dan else**
  *   Percabangann digunakan untuk memeriksa kondisi tertentu dan menjalankan blok kode yang sesuai.

*   Mengecek apakah suhu kurang dari atau sama dengan 90°C.
*   Jika benar, maka program menampilkan :
**STATUS: Normal**

  **b. elif suhu > 90 and suhu < 100:**
*   Kata elif berarti else if, digunakan untuk kondisi kedua. kondisi ini hanya akan dicek jika kondisi pertama salah.
*   Mengecek apakah suhu lebih dari 90°C tetapi kurang dari 100°C. jika benar, maka tampil :
**STATUS: Warning**

  **c. else:**
*   Baris ini akan dijalankan jika semua kondisi diatas tidak terpenuhi.

```python
# Soal 5

# Inisialisasi suhu awal
suhu_saat_ini = 60
target_suhu = 85

# Looping while untuk menaikkan suhu hingga mencapai target
while suhu_saat_ini < 85:
    print(f"Suhu saat ini: {suhu_saat_ini}°C")
    suhu_saat_ini += 5  # Menambah suhu 5 derajat setiap iterasi

# Cetak suhu akhir ketika target tercapai
print(f"Target suhu tercapai! Suhu akhir: {suhu_saat_ini}°C")
```

```python
# Soal 7

# Program klasifikasi beban listrik

# Input kapasitas maksimum dan beban terpakai
kapasitas_maks = float(input("Masukkan kapasitas maksimum (Watt): "))
beban_terpakai = float(input("Masukkan beban terpakai (Watt): "))

# Hitung persentase pemakaian
persentase = (beban_terpakai / kapasitas_maks) * 100

# Status beban
if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"


print(f"Persentase pemakaian: {persentase:.2f}%")
print(f"Status beban: {status}")
```

 **Penjelasan soal 7**

 **1. Kapasitas_maks = float(input("Masukkan kapasitas maksimum (Watt): "))**

*   Baris ini meminta pengguna untuk memasukkan kapasitas maksimum dari beban listrik dalam satuan Watt.
*   Fungsi input() digunakan untuk menerima data dari keyboard.
*   Fungsi float() mengubah data yang dimasukkan dari string menjadi bilangan desimal (float) agar bisa dihitung.

**2. Beban_terpakai = float(input("Masukkan beban terpakai (Watt): "))**

*   Baris ini meminta pengguna memasukkan beban listrik yang sedang digunakan.
*   Sama seperti sebelumnya, float() digunakan agar nilai bisa diproses dalam bentuk angka desimal.

**3. Persentase = (beban_terpakai / kapasitas_maks) * 100**

*   Bagian ini menghitung persentase pemakaian daya listrik dengan rumus:
Persentase = (beban terpakai / kapasitas maksimum) x 100%.
*   Hasilnya berupa nilai persentase (%) yang menunjukkan seberapa besar daya digunakan dibandingkan kapasitas total.

**4. Percabangan if – elif – else**

Digunakan untuk menentukan status beban listrik berdasarkan hasil persentase pemakaian.


1.   Jika persentase ≤ 60
*   Kondisi ini mengecek apakah penggunaan listrik masih ringan (≤ 60%).
*   Jika benar, maka variabel status diisi dengan teks "Beban Ringan".

2.   Jika persentase > 60 dan ≤ 85
*   Kondisi ini dijalankan jika pemakaian lebih dari 60% tetapi tidak lebih dari 85%.
*   Artinya, beban listrik berada pada kategori sedang.

3.   Jika persentase > 85
*   Jika kedua kondisi sebelumnya tidak terpenuhi, berarti persentase lebih dari 85%.
*   Maka, sistem akan memberikan status “Beban Berat”, yang artinya penggunaan listrik sudah mendekati atau melewati batas aman.

**5.   Output hasil**

 Menampilkan hasil akhir ke layar:
   *   Nilai persentase pemakaian ditampilkan dengan dua angka di belakang koma (.2f).
   *   Menampilkan status beban sesuai hasil percabangan sebelumnya.


































```python
# Challenge

# Fungsi untuk menghitung energi
def hitung_energi(power, jam):
    return power * jam  # dalam Wh

# Input jumlah perangkat
n = int(input("Masukkan jumlah perangkat listrik: "))

energi_total = 0
perangkat = 1

# Perulangan while untuk input data tiap perangkat
while perangkat <= n:
    print(f"Perangkat {perangkat}:")
    power = float(input("  Daya (Watt): "))
    jam = float(input("  Lama penggunaan per hari (jam): "))
    energi = hitung_energi(power, jam)
    print(f"  Energi yang digunakan per hari: {energi:.2f} Wh")
    energi_total += energi
    perangkat += 1

# Perulangan for untuk simulasi penggunaan selama 7 hari
total_seminggu = 0
for hari in range(1, 8):
    print(f"Hari ke-{hari}, energi digunakan: {energi_total:.2f} Wh")
    total_seminggu += energi_total

# Percabangan untuk mengkategorikan konsumsi
if total_seminggu < 1000:
    print("Kategori konsumsi: Hemat")
elif total_seminggu < 5000:
    print("Kategori konsumsi: Sedang")
else:
    print("Kategori konsumsi: Boros")

print(f"Total energi digunakan selama seminggu: {total_seminggu:.2f} Wh")
```

**Penjelasan Challenge**

**1. Fungsi hitung_energi(power, jam):**

*   Fungsi ini menerima dua parameter: daya (power) dalam watt dan lama penggunaan per hari (jam).
*   Fungsi mengembalikan hasil perkalian power dan jam, yaitu energi yang digunakan dalam watt-jam (Wh).

**2. nput jumlah perangkat listrik (n):**

*   Program meminta pengguna memasukkan jumlah perangkat listrik yang ingin dihitung konsumsi energinya.
*   Input ini dikonversi menjadi tipe integer.

**3. Inisialisasi variabel:**

*   energi_total diinisialisasi nol untuk menyimpan total energi semua perangkat.
*   Variabel perangkat digunakan sebagai penghitung loop mulai dari 1.

**4. Perulangan while untuk input data tiap perangkat:**

*   Selama perangkat kurang atau sama dengan n, maka program akan meminta input daya dan lama pemakaian perangkat tersebut.
*   Data input daya dan jam dikonversi ke tipe float.
*   Memanggil fungsi hitung_energi untuk menghitung energi yang dipakai perangkat tersebut per hari, dan hasilnya ditampilkan.
*   Energi tiap perangkat dijumlahkan ke energi_total.
*   Variabel perangkat dinaikkan 1 untuk memproses perangkat berikutnya.

**5. Perulangan for untuk simulasi penggunaan selama 7 hari:**

*   Program melakukan loop sebanyak 7 kali (untuk 7 hari).
*   Setiap hari, menampilkan berapa energi yang digunakan (energi_total per hari).
*   Menambahkan energi harian ke variabel total_seminggu.

**6. Percabangan untuk mengkategorikan konsumsi energi:**

*   Jika total energi selama seminggu kurang dari 1000 Wh, output "Kategori konsumsi: Hemat".
*   Jika antara 1000 hingga kurang dari 5000 Wh, output "Kategori konsumsi: Sedang".
*   Jika sama dengan atau lebih dari 5000 Wh, output "Kategori konsumsi: Boros".

**7. Output akhir:**

*   Menampilkan total energi yang digunakan selama seminggu.
