```python
#untuk memudahkan replacing maka di buatkan mapping jurusan sesuai dengan yg di inginkan, dengan kata-kata yg mirip di masukan ke jurusan yg seusai

mapping_jurusan = {
    "energi": "Pembangkit Energi",
    " energi ": "Pembangkit Energi",
    "ENERGI": "Pembangkit Energi",
    "EnErGi": "Pembangkit Energi",
    "eng": "Pembangkit Energi",
    "teknik": "Pembangkit Energi",
    "Teknik ": "Pembangkit Energi",
    "teknik ": "Pembangkit Energi",
    "manajemenenergi": "Manajemen Energi",
    "manajemen energi": "Manajemen Energi"
}
```

**Penjelasan:**

Kode di atas digunakan untuk membuat sebuah mapping atau peta perubahan nilai pada kolom jurusan untuk mempermudah proses normalisasi dan penyamaan data jurusan. Mapping tersebut disimpan dalam variabel mapping_jurusan berbentuk dictionary, di mana setiap kata atau variasi penulisan jurusan yang miripseperti "energi", "ENErGI", "eng", "teknik", maupun variasi dengan spasiakan diganti menjadi satu nama jurusan yang benar dan konsisten, yaitu "Pembangkit Energi". Begitu juga penulisan seperti "manajemenenergi" atau "manajemen energi" akan diseragamkan menjadi "Manajemen Energi". Mapping ini sangat berguna untuk mengatasi perbedaan penulisan yang tidak konsisten sehingga data menjadi lebih rapi, terstandarisasi, dan mudah dianalisis.

```python
tika_df["jurusan"] = tika_df["jurusan"].replace(mapping_jurusan)
```

**Penjelasan:**

kode tersebut digunakan untuk mengganti nilai-nilai pada kolom jurusan berdasarkan daftar mapping yang sudah dibuat sebelumnya. Fungsi replace() akan mencari seluruh kata atau variasi penulisan jurusan yang terdapat dalam dictionary mapping_jurusan, lalu menggantinya dengan nama jurusan yang sudah distandarisasi, seperti "Pembangkit Energi" atau "Manajemen Energi". Dengan perintah ini, seluruh data jurusan yang sebelumnya memiliki penulisan berbeda-beda menjadi lebih rapi, konsisten, dan seragam, sehingga memudahkan proses analisis dan pengelompokan data selanjutnya.
