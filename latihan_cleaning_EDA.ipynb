**Menangani inaccurate value**

sekarang kita akan mengatasi masalah inaccurate value pada kolom age. Sebagai awal, kita perlu melihat data baris data yang mengandung inaccurate value tersebut (baris dengan nilai age maksimum). Hal ini dilakukan menggunakan teknik filter seperti contoh kode berikut.

```python
customers_df[customers_df.age == customers_df.age.max()]
```

Kode `customers_df[customers_df.age == customers_df.age.max()]` digunakan untuk menampilkan **baris data pelanggan yang memiliki umur paling tua** di dalam DataFrame `customers_df`. Bagian `customers_df.age.max()` mencari nilai umur tertinggi, lalu ekspresi `customers_df.age == customers_df.age.max()` menghasilkan *True* untuk baris yang umur-nya sama dengan nilai maksimum tersebut. Hasilnya adalah tabel yang berisi satu atau lebih pelanggan dengan usia tertua dalam dataset.

Berdasarkan data tersebut, kita bisa berasumsi bahwa **inaccurate value** tersebut terjadi karena **human error** sehingga kelebihan memasukkan nilai nol. Oleh karena itu, gantilah dengan **nilai 70** . Proses ini dilakukan dengan memanfaatkan **method** `replace()`seperti contoh berikut.

```python
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

Kode `customers_df.age.replace(customers_df.age.max(), 70, inplace=True)` digunakan untuk **mengganti nilai umur tertentu di dalam kolom `age`** pada DataFrame `customers_df`.
```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

Kode tersebut digunakan untuk menggabungkan dua DataFrame, yaitu **orders_df** dan **customers_df**, menjadi satu tabel baru bernama **orders_customers_df**. Proses penggabungan dilakukan dengan fungsi **pd.merge()** dari pandas. Parameter **left_on="customer_id"** dan **right_on="customer_id"** berarti kedua tabel digabung berdasarkan kolom yang sama, yaitu *customer_id*. Metode **how="left"** membuat penggabungan model *left join*, artinya seluruh baris dari **orders_df** tetap muncul meskipun mungkin tidak memiliki pasangan data pada **customers_df**. Jika ada kecocokan customer_id, maka data pelanggan akan ditambahkan ke baris pesanan yang sesuai; jika tidak ada, kolom dari tabel customers akan terisi nilai NaN. Terakhir, **orders_customers_df.head()** digunakan untuk menampilkan beberapa baris awal hasil gabungan tersebut.


Pada kode di atas, kita melakukan proses merge dengan metode left join.  metode ini memungkinkan kita untuk mengambil semua nilai dari tabel kiri serta nilai yang bersesuaian dengan tabel kanan.
